1.4 Mensuração envolve MEDIDAS DE TENDÊNCIA CENTRAL

1.4.1 MÉDIA

```r
media_part <- aggregate(x = df$QT_VOTOS_NOMINAIS, by = list(df$SG_PARTIDO), FUN = mean)                          

media_part[order(-media_part$x),]
```

1.4.2 MEDIANA

```r
mediana_part <- aggregate(x = df$QT_VOTOS_NOMINAIS, by = list(df$SG_PARTIDO), FUN = median)                          

mediana_part[order(-mediana_part$x),]
```

1.4.3 MODA

```r
mode <- function(codes){
  which.max(tabulate(codes))
}

mds <- df %>%
  group_by(SG_PARTIDO) %>%
  summarise(mode = mode(QT_VOTOS_NOMINAIS))
mds[order(-mds$mode),]
```

1.5 Mensuração envolve MEDIDAS DE DISPERSÃO

1.5.1 VARIÂNCIA

```r
variance_part <- aggregate(x = df$QT_VOTOS_NOMINAIS, by = list(df$SG_PARTIDO), FUN = var)                          

variance_part[order(-variance_part$x),]
```

1.5.2 DESVIO PADRÃO

```r
library(forcats)
# DESVIOS PADRÃO POR PARTIDOS:
sd_part <- aggregate(x = df$QT_VOTOS_NOMINAIS, by = list(df$SG_PARTIDO), FUN = sd)                          
sd_part[order(-sd_part$x),]

# DESVIO PADRÃO DO CONJUNTO TOTAL DE VOTOS (com gráfico):
sum_partid <- aggregate(x = df$QT_VOTOS_NOMINAIS, by = list(df$SG_PARTIDO), FUN = sum)                          


media<-mean(sum_partid$x)
sdev<-sd(sum_partid$x)
sum_partid %>% 
mutate(Group.1 = fct_reorder(Group.1, x)) %>%
  ggplot(aes(x=Group.1, y=x)) +
  geom_bar(aes(x=Group.1, y=x), stat="identity", fill="forestgreen", alpha=0.5) +
  geom_errorbar( aes(x=Group.1, ymin=media-sdev, ymax=media+sdev), width=0.4, colour="orange", alpha=0.9, size=0.5) +
  coord_flip() +
  ggtitle("Desvio Padrão do conjunto de dados sobreposto")
```

1.5.3 COEFICIENTE DE VARIAÇÃO

```r
install.packages("EnvStats")
library(EnvStats)

cv_part <- aggregate(x = df$QT_VOTOS_NOMINAIS, by = list(df$SG_PARTIDO), FUN = cv)                          

cv_part[order(-cv_part$x),]
```
